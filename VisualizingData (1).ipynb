```python
def cal_per(col):
  percentage = col.value_counts(normalize=True).rename_axis(col.name).reset_index(name = 'percentage')
  return percentage
```

```python
from matplotlib.cbook import file_requires_unicode
def pl_fig(col, title1, title2):
  plt.figure(figsize = (15,4))
  # plt.subplot(1,2,1)
  per = cal_per(col)
  # plt.title(title1)
  mh_fdf.groupby(col).sum().plot(kind='pie', subplots=True,colormap='Set2', figsize = (10, 5), autopct = '%1.1f%%')
  plt.legend(bbox_to_anchor=(1, 0.5), loc="center left")

  plt.show()
  # sns.barplot(x = col.name, y = 'percentage', data = per,palette='Set1')
  # plt.subplot(1,2,2)
  plt.title(title2)
  sns.countplot(x = col, y=None , hue=mh_fdf['treatment'],palette='Set2')
  plt.xticks(rotation=45)
  # folder_path = '/content/drive/MyDrive/MentalHealth/output images'
  # plt.savefig(f'{folder_path}/{col.name}')
  plt.show()
```

```python
title1 = 'Employement type of respondents'
title2 = 'Employement type of respondents who are seeking treatment'
pl_fig(mh_fdf['self_employed'],title1, title2)
```
